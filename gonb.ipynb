```go

type CitirocProbeRegister struct {
	fastShaperOuput    [32]bool
	slowShaperLGOutput [32]bool
	peakSensingLG      [32]bool
	slowShaperHGOutput [32]bool
	peakSensingHG      [32]bool
	preampHGOutput     [32]bool
	preampLGOutput     [32]bool
	inputDAC           [32]bool
}
```

```go
%% 

initialValues := [32]bool{false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false}

probeRegister := CitirocProbeRegister {
    fastShaperOuput    : initialValues,
    slowShaperLGOutput : initialValues,
    peakSensingLG      : initialValues,
    slowShaperHGOutput : initialValues,
    peakSensingHG      : initialValues,
    preampHGOutput     : initialValues,
    preampLGOutput     : initialValues,
    inputDAC           : initialValues,
}

log.Println(probeRegister)
```

```go
func convertToBinary(values [32]bool) uint32{
    var result uint32 = 0
    for i:=0; i<len(values); i++{
        if values[i] {
            result = result | (1 << (31-i))
        }
    }
    log.Println(result)
    return result
}

func probeRegisterToBinary(register CitirocProbeRegister) []byte{
    fastShaperOuput := convertToBinary(register.fastShaperOuput)
    slowShaperLGOutput := convertToBinary(register.slowShaperLGOutput)
    peakSensingLG := convertToBinary(register.peakSensingLG)
    slowShaperHGOutput := convertToBinary(register.slowShaperHGOutput)
    peakSensingHG := convertToBinary(register.peakSensingHG)
    preampHGOutput := convertToBinary(register.preampHGOutput)
    preampLGOutput := convertToBinary(register.preampLGOutput)
    inputDAC := convertToBinary(register.inputDAC)
    
    result := uint32ToByteArray(fastShaperOuput)
    result = append(result, uint32ToByteArray(slowShaperLGOutput)...)
    result = append(result, uint32ToByteArray(peakSensingLG)...)
    result = append(result, uint32ToByteArray(slowShaperHGOutput)...)
    result = append(result, uint32ToByteArray(peakSensingHG)...)
    result = append(result, uint32ToByteArray(preampHGOutput)...)
    result = append(result, uint32ToByteArray(preampLGOutput)...)
    result = append(result, uint32ToByteArray(inputDAC)...)
    return result
}

func main() {
    initialValues := [32]bool{
        true,false,false,false,false,false,false,false,false,
        false,false,false,false,false,false,false,false,false,
        false,false,false,false,false,false,false,false,false,
        false,false,false,false,true}
    convertToBinary(initialValues)
    
    probeRegister := CitirocProbeRegister {
        fastShaperOuput    : initialValues,
        slowShaperLGOutput : initialValues,
        peakSensingLG      : initialValues,
        slowShaperHGOutput : initialValues,
        peakSensingHG      : initialValues,
        preampHGOutput     : initialValues,
        preampLGOutput     : initialValues,
        inputDAC           : initialValues,
    }
    
    result := probeRegisterToBinary(probeRegister)
    log.Println(result)
}
```

```go
func uint32ToByteArray(value uint32) []byte{
    array := make([]byte, 4)
    binary.LittleEndian.PutUint32(array, value)
    return array
}
```

```go
func binaryToProbeRegister(data []byte) CitirocProbeRegister{
    result := CitirocProbeRegister {
        fastShaperOuput    : binary.LittleEndian.Uint32(data[0:4]),
        slowShaperLGOutput : binary.LittleEndian.Uint32(data[4:8]),
        peakSensingLG      : binary.LittleEndian.Uint32(data[8:12]),
        slowShaperHGOutput : binary.LittleEndian.Uint32(data[12:16]),
        peakSensingHG      : binary.LittleEndian.Uint32(data[16:20]),
        preampHGOutput     : binary.LittleEndian.Uint32(data[20:24]),
        preampLGOutput     : binary.LittleEndian.Uint32(data[24:28]),
        inputDAC           : binary.LittleEndian.Uint32(data[28:32]),
    }
    return result
}
```
