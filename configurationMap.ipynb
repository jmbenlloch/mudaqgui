```go
func makeRange(start, end int) []int {
    a := make([]int, end-start+1)
    for i := range a {
        a[i] = start + i
    }
    return a
}

func makeRangeReverse(end, start int) []int {
    a := make([]int, end-start+1)
    for i := range a {
        a[i] = end - i
    }
    return a
}

func makeRangeReverseStep(end, start, step int) []int {
    a := make([]int, (end-start)/step +1)
    for i := range a {
        a[i] = end - step*i
    }
    return a
}

func makeRangeStep(start, end, step int) []int {
    a := make([]int, (end-start)/step +1)
    for i := range a {
        a[i] = start + step*i
    }
    return a
}
```

```go
%%
citirocProbeBitPosition := map[string][]int{
    "fastShaperOuput": makeRange(0, 31),
    "slowShaperLGOutput": makeRange(32, 64),
    "peakSensingLG": makeRange(64, 95),
    "slowShaperHGOutput" : makeRange(96, 127),
    "peakSensingHG": makeRange(128, 159),
    "preampHGOutput": makeRangeStep(160, 223, 2),
    "preampLGOutput": makeRangeStep(161, 223, 2),
    "inputDAC": makeRange(224, 255),
}


initialValues := [32]bool{true,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false,false}

probeRegister := map[string]any {
    "fastShaperOuput"    : initialValues,
    "slowShaperLGOutput" : initialValues,
    "peakSensingLG"      : initialValues,
    "slowShaperHGOutput" : initialValues,
    "peakSensingHG"      : initialValues,
    "preampHGOutput"     : initialValues,
    "preampLGOutput"     : initialValues,
    "inputDAC"           : initialValues,
}

//fmt.Println(probeRegister)


bit_length := 256
bytearray := make([]byte, bit_length / 8)

for key, values := range probeRegister{
    fmt.Println(key, values)
    fmt.Println(key, citirocProbeBitPosition[key])
    
    v := reflect.ValueOf(values)
    fmt.Println(v.Kind())
    fmt.Println(v)
    
    switch v.Kind() {
        case reflect.Array:
            fmt.Println("hu-array!")
            fmt.Println(v.Index(0))
            for i:=0; i<v.Len(); i++{
                position := citirocProbeBitPosition[key][i]
                fmt.Println(v.Index(i), position)
                if v.Index(i).Bool() {
                    arrayPosition := position / 8
                    bytearray[arrayPosition] = bytearray[arrayPosition] | (1 << (position % 8))   
                }
            }
        case reflect.Int:
            fmt.Println("integer!")
            fmt.Println(v.Int() + 3)
        case reflect.Bool:
            fmt.Println("booelan!")
            fmt.Println(v.Bool())
    }
}

fmt.Println(bytearray)
```
